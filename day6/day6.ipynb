1. Define a class Employee to store information of an employee (empNo, name, department,
basicPay, DA, HRA and grossSalary). Write a constructor which will take as input the
empNo, name, department, basicPay for several employees. The program will calculate the
DA, HRA and total for each employee and define a method to display the details of the
employee having the highest gross salary.

```python
class Employee:
  def __init__(self, empNo, name, dept, basicPay):
    self.empNo = empNo
    self.name = name
    self.dept = dept
    self.basicPay = basicPay

  def calc(self):
    self.da = self.basicPay * 0.11
    self.hra = self.basicPay * 0.06
    self.gross = self.basicPay + self.da + self.hra

  def getGrossVal(self):
    self.calc()
    return self.gross

  def display(self):
    print("Employee Number: ", self.empNo)
    print("Employee Name: ", self.name)
    print("Department Name: ", self.dept)
    print("Basic Pay: ", self.basicPay)

def getEmpData(self):
  empNo = input("Enter your Employee Number: ")
  name = input("Enter your Employee Name: ")
  dept = input("Enter your Department Name: ")
  basicPay = float(input("Enter your Basic Pay: "))
  return Employee(empNo, name, dept, basicPay)

def main():
  li = []
  n = int(input("Enter the number of employees: "))
  h = None
  for i in range(n):
    print(f"Enter Details of Employee No {i+1} : ")
    li.append(getEmpData(i))

    if not h:
      h=li[i]
    elif h.getGrossVal() < li[i].getGrossVal():
      h=li[i]
    print()

  h.display()

if __name__ == '__main__':
  main()
```

3. import re and write scripts to achieve the following:
(i) accept a mobile number as input, and if it starts with 00, replace the 00 with 91
(ii) takes a word as input and satisfy the requirements that it should be of 5 letters,
starts with ‘c’ and ends with ‘r’
(iii) assume a product cataloging system has the following requirements:

first symbol: an uppercase character
second symbol: a digit
third symbol: the special character @
fourth symbol: a lowercase character
write a script that accepts an id and checks whether it is in correct format or not
(iv) check whether a given mail id is in valid format or not.

```python
import re

# (i)
x = input("Mobile No: ")
print(re.sub("\A00","91",x)) if re.search("^00", x) else print("WRONG INPUT!!!")

# (ii)
y = input("Enter your word: ")
print("The word starts with c & r and is of 5 letters") if re.search("^c...r$", y) else print("The word doesn't satisfy the requirments!!!")

# (iii)
z = input("Enter an id: ")
print("The ID is in the Correct Format") if re.search("^[A-Z]+[0-9]+[@]+[a-z]$",z) else print("The ID is in the Incorrect Format!!!")

# (iv)
k = input("Enter your email: ")
print("The E-mail is valid") if re.fullmatch(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b",k) else print("The E-mail is invalid!!!")
```

method display_area() to compute and display the area of figures. Redefine this method in
the derived classes to suit their requirements. Use these three classes to design a program
that will accept dimensions of a triangle and rectangle and display the area.

```python
class Shape:
  def __init__(self, figure):
    self.figure = figure

    def display_area(self):
      pass

class Triangle(Shape):
  def __init__(self,b,h):
    super().__init__("Triangle")
    self.b = b
    self.h = h

  def display_area(self):
    area = 0.5 * self.b * self.h
    print(f"The area of {self.figure} is {area}.")

class Rectangle(Shape):
  def __init__(self,l,b):
    super().__init__("Rectangle")
    self.l = l
    self.b = b
    
  def display_area(self):
    area = self.l * self.b
    print(f"The are of {self.figure} is {area}.")

def main():
  tr = Triangle(20,10)
  tr.display_area()

  rec = Rectangle(40,20)
  rec.display_area()


if __name__ == '__main__':
  main()
```
